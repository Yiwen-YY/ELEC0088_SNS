```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tmin = pd.read_csv('tmin_forecast.csv')
tmin = df_tmin['tmin']
date = df_tmin['date']
```

```python
df_tmax = pd.read_csv('tmax_forecast.csv')
tmax = df_tmax['tmax']
```

```python
sns.set()
plt.rcParams['figure.figsize'] = [12, 4]
plt.plot(date, tmin, label="Minimum temperature", color="b") 
plt.plot(date, tmax, label='Maximum temperature', color='green') 
 
plt.xlabel('Date')
plt.ylabel(r'Temperature:Units(' + chr(176) + 'C)')
plt.xticks(rotation=90)
plt.legend()
plt.title('Time Series Forecasting for One Month')
plt.grid(True)
```
